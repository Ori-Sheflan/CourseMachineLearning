#### a. 
Let $K,L$ be two kernels (operating on the same space)  
let $\alpha,\beta$ be two positive scalars.   
Prove that $\alpha K + \beta L$ is a kernel.  


<u>Answer:</u>

$K(x,y)$ and $L(x,y)$ are kernels $\Rightarrow$ mapping functions $\phi_k,\phi_L \Rightarrow K(x,y) = \phi_k(x) \cdot \phi_k(y) \space\space , \space\space L(x,y) = \phi_L(x) \cdot \phi_L(y)$
  
And we will prove that $A(x,y) = \alpha k(x,y) + \beta L(x,y)$ is a kernel.

$
A(x,y) =   \\
= \alpha K(x,y) + \beta L(x,y) = \\
= \alpha \phi_K(x)\phi_K(y) + \beta \phi_L(x)\phi_L(y) = \\
= \alpha \bigg( \phi_{K,1}(x),\phi_{K,2}(x),...,\phi_{K,n}(x) \bigg)\cdot\bigg( \phi_{K,1}(y),\phi_{K,2}(y),...,\phi_{K,n}(y) \bigg) + \beta \bigg( \phi_{L,1}(x),\phi_{L,2}(x),...,\phi_{L,n}(x) \bigg)\cdot\bigg( \phi_{L,1}(y),\phi_{L,2}(y),...,\phi_{L,n}(y) \bigg) = \\
= \bigg( \sqrt{\alpha}\phi_{K,1}(x),...,\sqrt{\alpha}\phi_{K,n}(x) \bigg)\cdot\bigg( \sqrt{\alpha}\phi_{K,1}(y),...,\sqrt{\alpha}\phi_{K,n}(y) \bigg) +  \bigg( \sqrt{\beta}\phi_{L,1}(x),...,\sqrt{\beta}\phi_{L,n}(x) \bigg)\cdot\bigg( \sqrt{\beta}\phi_{L,1}(y),...,\sqrt{\beta}\phi_{L,n}(y) \bigg) = \\
= \bigg( \sqrt{\alpha}\phi_{K,1}(x),...,\sqrt{\alpha}\phi_{K,n}(x),\sqrt{\beta}\phi_{L,1}(x),...,\sqrt{\beta}\phi_{L,n}(x) \bigg) \cdot \bigg( \sqrt{\alpha}\phi_{K,1}(y),...,\sqrt{\alpha}\phi_{K,n}(y),\sqrt{\beta}\phi_{L,1}(y),...,\sqrt{\beta}\phi_{L,n}(y) \bigg) = \\
\text{And let's define a mapping function: } \space \phi(x) = \bigg( \sqrt\alpha\phi_{K,1}(x),\sqrt\alpha\phi_{K,2}(x),...,\sqrt\alpha\phi_{K,n}(x),\sqrt\beta\phi_{L,1}(x),\sqrt\beta\phi_{L,2}(x),...,\sqrt\beta\phi_{L,n}(x) \bigg) \\
\text{ then we can conclude:  } A(x,y) = \phi(x) \cdot \phi(y) \text{ therefore, A(x,y) is a kernel}
$
